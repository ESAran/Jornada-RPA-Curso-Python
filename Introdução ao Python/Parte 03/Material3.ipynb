##  ARQUIVOS DE TEXTO
------

Existem diferentes modos de abertura de arquivos:

"r" - read:  o valor padro, abre o arquivo para leitura e apresenta erro se o arquivo no existir.

"a" - append - abre o arquivo para adicionar contedo e cria um novo se o arquivo no existir.

"w" - write - abre o arquivo para escrita, sobrescreve se o arquivo j existir e criar um novo se o arquivo no existir.

"x" - create - cria um arquivo especfico e retorna um erro caso o arquivo j existir.

Podemos tambm definir se estamos lidando com um arquivo de texto ou arquivo binrio.

"t" - text:  o valor padro, modo de texto.

"b" - binary: modo binrio, por exemplo: images, arquivos PDF e etc.

```
with open("arquivo.txt", "w") as arq:
    arq.write("Estamos escrevendo a primeira linha do arquivo\n")
    arq.write("Vejam s, esta  a segunda linha do arquivo\n")
    arq.close()
```

```
arq = open("arquivo.txt", "r")

# leitura de uma linha do arquivo
# print(arq.readline())

# leitura de todas as linhas do arquivo e coloca em uma lista
print(arq.readlines())
arq.close()
```

```
with open("arquivo.txt", "a") as arq:
    arq.writelines(["Este  um terceiro contedo\n", "Este  o quarto contedo\n"])
```

```
print("Nome: " + arq.name)
print("Modo: " + arq.mode)
print("Tipo: " + str(type(arq)))
```

```
arq = open("arquivo.txt", "r")
# indica o ponto onde o cursor de movimentao do arquivo se encontra
print("Est em: %s" % arq.tell())
print(arq.readlines())
print("Est em: %s" % arq.tell())
arq.seek(0)
print("Est em: %s" % arq.tell())
```

```
with open("arquivo.txt", "r") as arq:
    # delimita a quantidade de caracteres a serem lidos
    print(arq.read(4))
    print(arq.tell())
    print(arq.read(6))
```

```
with open("arquivo.txt", "r") as arq:
    linhas = arq.readlines()
    for linha in linhas:
        print(linha)
```

### &nbsp;&nbsp;ESTRUTURAS DE DADOS

As estruturas de dados primrias no Pandas so implementadas em duas classes:

* **`DataFrame`**: uma estrutura tal qual uma tabela relacional com linhas e colunas.
* **`Series`**: que contm uma nica coluna

Assim, um `DataFrame` contm uma ou mais `Series` e um nome para cada `Series`.

A estrutura de DataFrame  bastante comum na abstrao para manipulao de dados e tambm  implementada em no framework [Spark](https://spark.apache.org/) e na Linguagem [R](https://www.r-project.org/about.html).

Objetos `DataFrame` podem ser criados passando um dicionrios mapeando os nomes das colunas para suas respectivas `Series`. Se a `Series` no coincide no tamanho, valores faltantes so preenchidos com [NA/NaN](http://pandas.pydata.org/pandas-docs/stable/missing_data.html).

#### Sries

Sries no estruturas unidimensionais que podem ter ndices numricos ou em outros formatos.

Podemos usar a estrutura de dados `Series` passando os valores em uma lista.
O segundo parmetro `index` indica qual deve ser o identificar de ndice para cada elemento. Caso no seja designados valores para o ndice, os valores sero nmeros em ordem crescente iniciando por 0.

\

```
s = pd.Series(["Azul", "Roxo", "Laranja"], index=[0, 2, 4])
display(s)
```

Para saber a quantidade de valores em uma srie usa-se `size`:

```
s.size
```

Para retornar um array com os valores contidos na srie, usa-se `values`:

```
s.values
```

Podemos acessar um valor diretamente pelo seu ndice com `at`:

```
s[2]
```

Estrutura para buscar todos `findall`

```
s.str.findall("Roxo")
```

Para exibio da lista, podemos adicionar elementos depois com `add()` e antes com `radd()`. Esses elementos so apenas para visualizao.

```
s.add("!").radd("#")
```

Para adicionar novos valores em uma srie o recomendado  utilizar a funo do pandas `concat`. Ela aceita apenas a concatenao de uma srie em outra, portanto deve-se passar como primeiro parmetro a srie atual e como segundo parmetro a nova srie a ser agregada.

```
s = pd.concat([s, pd.Series(["Amarelo", "Verde"], index=[6, 8])])
```

```
display(s)
```

Se no estipularmos os valores de ndice, podemos passar como terceiro  parmetro `ignore_index=True` para que seja estruturada a indexao adequadamente.

```
s1 = pd.Series(["Uva", "Maa", "Morango"])
display(s1)

s2 = pd.Series(["Banana", "Morango"])
display(s2)

sn = pd.concat([s1, s2], ignore_index=True)
display(sn)
```

Ao usar `is_unique` avalia se os valores todos so nicos. Neste caso ir retornar `False`, pois temos o valor "morango" repetido.

```
sn.is_unique
```

Para retornar em um array apenas os valores nicos, pode-se usar `unique()`

```
sn.unique()
```

Podemos reindexar os ndices caso seja necessrio. Na srie `s` temos atualmente os ndices 0, 2, 4, 6 e 8, se quisermos reindexar para que a srie tenha 10 valores, teremos que definnir quais so os valores faltantes.

Assim, existem opes para o mtodo de completude de dados:
`ffill`: ltimo valor vlido;
`bfill`: prximo valor vlido;
`nearest`: o valor mais prximos;  

Ainda, pode-se estabelecer um valor usando `fill_value` (por exemplo: `fill_value="VAZIO"`)



```
s.reindex(range(10), method="nearest") # fill_value="NULL"
```

#### DataFrame

Um DataFrame  um objeto que representa uma tabela e contm uma coleo ordenada de colunas em que cada uma pode ter um tipo de valor diferente. Possui um ndice tanto para linha quanto para coluna, podendo ser imaginado como um dicionrio de sries onde todos compartilham o mesmo ndice (MCKINNEY, 2018, p. 117).

Para criar um dataframe devemos usar a funo `DataFrame` da biblioteca Pandas, passando como parmetro um dicionrio de dados.

```
df = pd.DataFrame({"nome": ["Aline", "Pedro", "Marina", "Jos"],
                   "idade": [39, 35, 28, 45],
                   "pet": ["gato", "cachorro", "papagaio", "peixe"]})
```

```
display(df)
```

Para exibir as colunas de um DataFrame usa-se `columns`:

```
df.columns
```

J para exibir valores de um DataFrame usa-se `values`:

```
df.values
```

Criando um nome para a estrutura de ndices do DataFrame com `index.name`:

```
df.index.name = "id"
display(df)
```

Para insero de um novo valor no DataFrame, devemos usar a funo da biblioteca Pandas `concat` passando como parmetro outro DataFrame a ser concatenado.

```
dados = pd.DataFrame({ "nome": ["Abelino"],
                       "idade":[67],
                       "pet": ["cachorro"] })
pd.concat([df, dados], ignore_index=True)
```

Caso seja necessrio realizar a transposio de um DataFrame (as linhas tornam-se colunas e vice-versa) usa-se `T` associado ao DataFrame. Perceba que a transposio  feita apenas em tempo de execuo, para usar a estrutura transporta  preciso associar a um novo DataFrame.

```
df.T
```

Uma forma interessante de visualizao de um DataFrame  o oferecido pela funo `stack()`. Ele apresenta a estrutura em um formato presenta multinvel.


```
df.stack()
```

Se for necessrio inserir uma nova coluna no DataFrame, podemos usar a funo `insert()`. Passamos por parmetro a posio onde a coluna dever ser inserida, o nome da coluna (label) e o valor padro a ser inserido.

```
df.insert(3, "status", "ativo")
```

```
df
```

Para retirar uma coluna, usa-se `drop()` passando por parmetro o nome da coluna.

```
df.drop("status", axis="columns")
```

Caso seja necessrio verificar se determinado valor de ndica existe, devemos usar `in` buscando na lista de ndices `index`:

```
6 in df.index
```

### &nbsp;&nbsp;MANIPULAO DE DADOS

#### Lendo um arquivo csv

Na maioria das vezes que vamos trabalhar com a biblioteca Pandas a nossa necessidade ser realizar a leitura e manipulao de um conjunto de dados que esteja em um arquivo, banco de dados, API ou qualquer outra fonte externa.

Assim, podemos ler fontes de dados em diferentes formatos e para isso o Pandas oferece uma conjunto de funes.

|Funo | Descrio
|:---:|:---|
| read_csv() | leitura de dados deliminados no formato CSV - Comma Separated Values (padro: vrgula)
| read_table() | leitura de dados delimitados utilizando tabulao ('\t')
| read_fwf() | leitura de dados delimitados com tamanho fixo de colunas
| read_clipboard() | leitura de dados da rea de transferncia
| read_excel() | leitura de dados tabulares de arquivos Microsoft Excel
| read_hdf() | leitura de arquivos HDF escritos com o Pandas
| read_html() | leitura de tabelas que se encontram em documentos HTML
| read_json() | leitura de dados na representao JSON - Javascript Object Notation
| read_msgpack() | leitura de dados codificados pelo Pandas no formato binrio MessagePack
| read_pickle() | leitura de objeto armazenado no formato pickle do Python
| read_sas() | leitura de dados de SAS armazenado em um fos formatos personalizados do sistema SAS
| read_sql() | leitura do resultado de uma consulta SQL (usando SQLAlchemy) na forma de um DataFrame do Pandas
| read_stata() | leitura de dados no formato de arquivo Stata
| read_feather() | leitura de dados no formato de arquivo binrio Feather

Certamente uma das mais populares das funes listadas acima  a read_csv() uma vez que o formato Comma Separated Values (CSV)  bastante utilizado para distribuio de datasets.

####Seleo dos dados

Em muitos casos teremos datasets com muitas colunas e nem sempre precisaremos utilizar todas elas. Carregar sempre todas as colunas pode ser oneroso em se tratando de processamento e at mesmo organizao da nossa prtica. Sendo assim, podemos determinar quais so as colunas que devem ser lidas e inseridas no DataFrame.

```
dfn = pd.DataFrame(df, columns=["host_name", "name", "neighbourhood", "room_type", "minimum_nights", "price", "number_of_reviews", "reviews_per_month", "last_review"])
display(dfn)
```

Se necessitarmos criar um DataFrame com apenas algumas linhas, podemos usar a funo `iloc` passando em uma lista os ndices dos registros necessrios. Podemos usar intervalos de valores como `iloc[10:30]` que ir retornar os registros dos ndices 10 at 29. Se os registros tiverem ndices em valores no numricos, podemos usar `loc()`.

```
df.iloc[[1, 2, 4, 10]]
```

#### Operaes em dados

Podemos fazer algumas operaes em dados como agrupamentos e ordenaes.

#### Agrupamentos

Usando a funo `groupby()` podemos passar a(s) coluna(s) que queremos agrupar. No caso a seguir, estamos contando (funo `count()`) os registros da coluna `name` agrupando pela coluna `neighbourhood`:

```
dfn.groupby("neighbourhood")["name"].count()
```

Caso seja mais de um valor, temos que passar como listas:

```
dfn.groupby(["neighbourhood", "room_type"])["name"].count()
```

Se quiser exibir apenas os registros de um determinado grupo, pode-se usar a funo `get_group()` passando por parmetro o grupo desejado.

```
dfn.groupby("neighbourhood").get_group("Copacabana")
```

Temos alguns objetos implcitos com informaes que podem ser teis sobre o agrupamento. O objeto `groups` mostra um dicionrio de dados completo com todos os ndices para cada registro de agrupamento realizado.

```
dfn.groupby("neighbourhood")["name"].groups
```

J o objeto `indices` apresenta um dicionrio com arrays de ndices para cada valor gerado pelo agrupamento.

```
dfn.groupby("neighbourhood")["name"].indices
```

Por fim, `ngroups` apresenta a quantidade de grupos criados pelo agrupamento.

```
dfn.groupby("neighbourhood")["name"].ngroups
```

Em caso da necessidade de agregao de mais de uma operao no agrupamento, podemos usar a funo `aggregate` indicando em uma lista quais operaes devem ser realizadas:

```
dfn.groupby("neighbourhood")["number_of_reviews"].aggregate(["count", "sum"])
```

#### Filtragem de dados

Podemos criar consultas de uma maneira muito semelhante ao que realizamos em consultas em bases de dados relacionais. A forma mais simples  criando o filtro dentro do DataFrame. No exemplo a seguir estamos selecionando todos os registros que o preo esteja entre 200 e 500 reais"

```
dfn[ (dfn["price"] >= 200) & (dfn["price"] <= 500) ]
```

Aqui uma consulta com mais critrios e o uso de operadores relacionais e lgicos. Importante perceber que para AND devemos usar o smvolo `&` e para OR o smbolo `|`. Neste caso estamos filtrando os registros que tenham bairro "Copacabana" E com preo maior ou igual a 1000 E que a quantidade mnima de noites seja 4 OU que tenha mais do que 10 reviews.

```
dfn[
    (dfn["neighbourhood"] == "Copacabana") &
    (dfn["price"] >= 1000) &
    ( ( dfn["minimum_nights"] == 4) | (dfn["number_of_reviews"] > 10) )
]
```

Podemos usar a funo `isin()` para avaliar se um valor consta no DataFrame. O retorno ser sempre verdadeiro ou falso. Ao inserimos como filtro do DataFrame teremos o retorno das linhas que atendem ao critrio.

```
dfn[dfn["host_name"].isin(["Claudia", "David"])]
```

Com a funo `contains()` tambm podemos usar uma estratgia semelhante ao "LIKE" das consultas em SQL nos bancos de dados relacionais. Assim, buscaremos por um termo no contedo de determinada coluna do DataFrame. O atributo `case=False` faz com que seja ignorada a diferena entre maisculas e minsculas.

```
dfn[dfn["name"].str.contains("Piscina", case=False).fillna(False)]
```

Para buscas simples com operadores relacionais e lgicos podemos usar tambm a funo `query()`

```
# dfn[ (dfn["neighbourhood"] == "Copacabana") & (dfn["price"] >=500) ]
dfn.query("neighbourhood == 'Copacabana' & price <= 500")
```

```
dfn.query("neighbourhood == 'Copacabana' & price <= 500")
```

#### Ordenao de dados

Para ordenao dos dados, podemos usar a funo `sort_values()` passando como parmetro o rtulo da coluna que guiar a sequncia.

```
dfn.sort_values("price")
```

Por padro os valores so apresentados em ordem crescente, assim se quiseremos que seja apresentado em ordem decrescente temos que usar o parmetro `ascending=False`.

```
dfn.sort_values("price", ascending=False)
```

Tambm podemos fazer a ordenao pelos ndices dos registros.

Podemos usar quantas colunas forem necessrias para estabelecer a ordenao, basta que seja indicado em formato de lista como parmetro da funo `sort_values()`.

```
dfn.query("neighbourhood == 'Copacabana' & price <= 500").sort_values(["minimum_nights", "price"])
```

#### Tratamento de dados

Para efetuar operaes de tratamento de dados nos DataFrames a biblioteca Pandas apresenta algumas funes.

Por exemplo, a funo `drop_duplicates()` que remove linhas com valores duplicados.

```
dfn.drop_duplicates()
```

A funo `isnull()` ou `isna()` que verifica se o registro contm valores fantantes (missing values).

```
dfn.isnull()
```

J a funo `dropna()` remove os registros com valores faltantes. Essa operao  realizada em tempo de execuo, por isso para registrar seus efeitos no DataFrame deve-se usar o parmetro `inplace=True` ou ento associar a um novo DataFrame.

```
dflimpo = dfn.dropna()
display(dflimpo)
```

Pode tambm preencher os dados faltantes com o valor desejado usando a funo `fillna()`:

```
dfn = dfn.fillna("NULL")
display(dfn)
```
