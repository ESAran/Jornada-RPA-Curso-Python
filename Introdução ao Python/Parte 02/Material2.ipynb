## &nbsp;&nbsp;MODULARIZAO

---

A modularidade trata da reutilizao de cdigo, abstrao de complexidade e facilitao de manutenibilidade. A construo de funes permitem o isolamento de blocos de cdigo que executam tarefas especficas, facilitando a leitura, a manuteno e o teste. Ao encapsular a lgica em funes, podemos evitar a redundncia, reduzir erros e garantir que as alteraes feitas em um ponto do cdigo sejam refletidas de maneira consistente em toda a aplicao. Alm disso, funes facilitam a colaborao e a escalabilidade, pois diferentes desenvolvedores podem trabalhar em funes distintas sem interferir no trabalho uns dos outros.

###Customizao de funes


Para criarmos funes que podero ser chamadas de acordo com a necessidade, usamos a palavra reserva ``` def ``` seguida do nome da funo e os sinais de parnteses para indicar a presena ou no de parmetros.


```python
def function_name(argument1, argument2,...,):
    """
    Description of function
    """

    return(return_value)
```



````python
def quadrado(x):
  return x ** 2

def multiplicacao(a, b):
  return a * b
````

### Parmetros e retorno em funes

Assim como em outras linguagens de programao podemos construir diferentes tipos de funes em termos de passagem de parmetros e existncia ou no de instrues de retorno.

Podemos passar quantos parmetros forem necessrios para uma funo:


````python
from datetime import datetime

# Declarao de funo que NO TEM parmetros e NO TEM retorno
def boas_vindas_fixo():
    print("Seja bem vindo(a)!")

# Declarao de funo que TEM parmetros e NO TEM retorno
def boas_vindas_dinamico(mensagem):
    print(mensagem)

# Declarao de funo que NO TEM parmetros e TEM RETORNO
def boas_vindas_com_horario():
    return "Seja bem vindo! Agora so " + datetime.now().strftime("%H:%M de %d/%m/%Y")

# Declarao de funo que TEM parmetros e TEM RETORNO
def boas_vindas_com_nome(nome):
    return "Seja bem vindo(a) " + nome + "!"
````

````python
# Chamada da funo que NO TEM parmetros e NO TEM retorno
boas_vindas_fixo()
````

````python
# Chamada da funo que TEM parmetros e NO TEM retorno
boas_vindas_dinamico("Ol! Seja muito bem vindo(a)! :)")
````

````python
# Declarao de funo que NO TEM parmetros e TEM RETORNO
mensagem_horario = boas_vindas_com_horario()
print("A mensagem retornada foi: " + mensagem_horario)
````

````python
# Chamada de funo que TEM parmetros e TEM RETORNO
mensagem_nome = boas_vindas_com_nome("Maria")
print("A mensagem retornada foi: " + mensagem_nome)
````

### Parmetros com valores padro

Podemos pr-definir valores padro para parmetros. Assim, se ao chamarmos uma funo e no passarmos valores para estes parmetros, automaticamente eles adotaro os valores padres definidos em suas declaraes.

Este primeiro exemplo apresenta quatro parmetros e no define valores padro. Logo ao realizar a chamada deve-se passar todos os valores para cada parmetro ou indica-los como vazios passando "".

````python
def dados_usuario(nome, cpf, email, pais):
    print("Nome: " + nome)
    print("CPF: " + cpf)
    print("E-mail: " + email)
    print("Pas: " + pais)

dados_usuario("Pedro Santos", "123.456.789-10", "pedro@email.com", "Brasil")
````

J no caso abaixo, note que os parmetros de email e pais esto com valores pr-definidos na declarao da funo. Logo, quando estes parmetros no forem passados na chamada da funo, eles vo automaticamente atribuir o valor que est como padro na funo.

Importante ressaltar que os parmetros que tem valores padro devem sempre estar definidos em sequncia ao final da lista de parmetros.

````python
def dados_usuario_padrao(nome, cpf, email = "No informado", pais = "Brasil"):
    print("Nome: " + nome)
    print("CPF: " + cpf)
    print("E-mail: " + email)
    print("Pas: " + pais)

dados_usuario_padrao("Pedro Santos", "123.456.789-10")
````

### Funes com estruturas de dados

Nas funes, alm de trabalhar com os valores nicos para cada parmetro, podemos tambm utilizar estruturas de dados o que amplia muito as possibilidades! Sendo assim, podemos passar como parmetros estruturas de dados como listas e dicionrios.

````python
# Recebe uma lista como parmetro e exibe todos os itens
def lista_compras(itens):
    for i in itens:
        print(i)
````

````python
# Define uma lista e passa esta estrutura como parmetro na chamada da funo
itens = ["Po", "Caf", "Manteiga", "Cereal"]
lista_compras(itens)
````

Da mesma forma, podemos tambm fazer com que o retorno de uma funo seja uma estrutura de dados, tanto de listas, quanto de dicionrios.

````python
import random

# Recebe uma lista e cria um dicionrio que  retornado como parmetro
def compras_quantidade(itens):
    completa = {}
    for i in itens:
        completa[i] = random.randint(2, 5)
    return completa
````

````python
# Chamada da funo que retorna um dicionrio armazenado em outra varivel
dicionario_completa = compras_quantidade(itens)
print("Dicionrio criado: ")
print(dicionario_completa)
````

Para apresentar as informaes de chave e de valor do dicionrio criado, podemos usar a funo `items()` e fazer a iterao em cada item.

````python
for k, v in dicionario_completa.items():
    print("Comprar " + str(v) + " unidades de " + k)
````

### Retorno de mltiplos valores

````python
def calculos(a, b):
    adicao = a+b
    subtracao = a-b
    divisao = a/b
    multiplicacao = a*b
    return(adicao, subtracao, divisao, multiplicacao)
````

````python
a, s, d, m = calculos(10, 3)
multiplicacao = calculos(10, 3)[3]

total = a + s + d + m
print("Multiplicacao resultou em: " + str(multiplicacao))
print("Adio resultou em: " + str(a))
print("Valor total da soma dos resultados: " + str(total))
````

### Lidando com excees


 um evento que ocorre durante a execuo de um programa quando h um desvio do fluxo normal das instrues programadas. Quando isso ocorre, cria-se ento um um objeto que representa um erro. Para previnir e/ou expressar esses erros corretamente existe o tratamento de excees. No Python existem dezenas de excees que podem ser mapeadas, voc pode encontrar uma lista completa delas em: https://docs.python.org/pt-br/3.12/library/exceptions.html#Exception

Assim, podemos criar o tratamento de exceo toda vez que verificarmos que h possibilidade que um erro acontea.

````python
12/0
````

````python
try:
    12 / 0
except ZeroDivisionError:
    print('No  possvel dividir por zero!')
````

````python
try:
    lista = [1, 2, 3]
    print(lista[4])try
except:
    print('Um erro ocorreu!')
````

Por padro, uma exceo gera um objeto que inclui informaes sobre o problema que causou o erro.

````python
try:
    lista = [1, 2, 3]
    print(lista[4])
except Exception as erro:
    print("Um erro ocorreu: " + str(erro.args[0]))
````

### Hierarquia de excees

As excees devem ser organizadas conforme uma hieraquia, isso significa que podemos ter mltiplas excees mapeadas.

````python
try:
   # 12/0
    lista = [1, 2, 3]
    print(lista[4])
except ZeroDivisionError:
    print("No  possvel dividir por zero ;)")
except Exception as erro:
    print("Um erro ocorreu!")
````
